```python
# 1️⃣ Import Library
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 2️⃣ Buat Data Contoh
data = pd.DataFrame({
    'Date': pd.date_range(start='2021-01-01', periods=10, freq='D'),
    'Cases': [10, 20, 15, 30, 25, 40, 35, 50, 45, 60],
    'Deaths': [1, 2, 1, 3, 2, 4, 3, 5, 4, 6],
    'Recovered': [5, 10, 7, 15, 12, 20, 17, 25, 22, 30]
})

# 3️⃣ Pastikan 'Date' dalam format datetime
data['Date'] = pd.to_datetime(data['Date'])

# 4️⃣ Plot Tren Kasus Harian
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Cases'], label='Kasus Harian', color='blue', marker='o')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.title('📊 Tren Kasus Harian COVID-19 di Indonesia')
plt.xticks(rotation=45)  # Putar label tanggal
plt.legend()
plt.grid(True)
plt.show()

# 5️⃣ Plot Tren Kematian Harian
plt.figure(figsize=(10, 6))
plt.bar(data['Date'], data['Deaths'], color='red', alpha=0.7, label='Kematian Harian')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kematian')
plt.title('📊 Tren Kematian Harian COVID-19 di Indonesia')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# 6️⃣ Plot Tren Kesembuhan Harian
plt.figure(figsize=(10, 6))
plt.bar(data['Date'], data['Recovered'], color='green', alpha=0.7, label='Kesembuhan Harian')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kesembuhan')
plt.title('📊 Tren Kesembuhan Harian COVID-19 di Indonesia')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
